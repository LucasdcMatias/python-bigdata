```python
# ler o arquivo venda.xlsx
import pandas as pd
```

```python
# ler o arquivo vendas.xlsx
df = pd.read_excel(r'/content/drive/MyDrive/Aula dia 21 08 - Valor Lojas/Vendas.xlsx')
# apresentar os dados
display(df)
```

```python
# quais os tipos de dados de cada coluna
df.dtypes
```

```python
# calculando o faturamento por loja - agrupando e rankeado os dados
faturamento = df[['ID Loja','Valor Final']].groupby('ID Loja').sum()
# ordernar pelo maior faturamento
faturamento = faturamento.sort_values(by = 'Valor Final' , ascending = False)

display(faturamento)
```

```python
# calculando a qantidade vendida por loja
quantidade = df[['ID Loja' , 'Quantidade']].groupby('ID Loja').sum()
quantidade = quantidade.sort_values(by = 'Quantidade', ascending = False)
display(quantidade)
```

```python
# calculando o ticket m√©dio por loja
ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns = {0: 'Ticket Medio'})
ticket_medio = ticket_medio.sort_values(by = 'Ticket Medio', ascending = False)
display(ticket_medio)
```
